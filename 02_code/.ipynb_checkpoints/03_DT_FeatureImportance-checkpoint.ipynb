```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as mpatches

from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
# Reading in the combined dataframe of combined AD and HV samples 
df = pd.read_csv('Total.csv')
```

```python
# Checking the head tail and shape of the dataframe
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
# Replacing the nan values with zero at the beginning of our dataframe 
df = df.fillna(0, axis=0)
```

```python
# Setting X and y values 
X = df.drop(columns=['Alzheimerâ€™s'])
y = df['Alzheimerâ€™s']
```

```python
# Instantiating our model
DecisionTreeClassifier()
```

```python
# Setting parameters and fitting the model
tree = DecisionTreeClassifier(max_depth=2)
tree.fit(X,y)
```

```python
# Creating a dataframe of features most important for classifying between our target groups 
pd.DataFrame({
    'feature': X.columns,
    'importances': tree.feature_importances_
}).sort_values('importances', ascending=False)
```

```python
# Filtering columns to find the full name of the columns 
# because they are partially cutoff
columns_filtered = df.filter(like='Keratin, type II cytoskeletal 3 OS=Homo sapien', axis=1)
```

```python
# Combing through possible matches 
print(columns_filtered)
```

## 1.) Visualization 

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x = target_df.iloc[:, 0]
y = target_df.iloc[:, 1]
z = target_df.iloc[:, 2]
labels = target_df.iloc[:, 3]
colors = labels.map({0: 'blue', 1: 'teal'})

ax.scatter(x, y, z, c=colors, s=60, alpha=0.8)
ax.set_xlabel(target_df.columns[0])
ax.set_ylabel(target_df.columns[1])
ax.set_zlabel(target_df.columns[2])
plt.title('3D Protein Abundance by Alzheimer\'s Status')

blue_patch = mpatches.Patch(color='blue', label='No Alzheimer\'s')
teal_patch = mpatches.Patch(color='teal', label='Alzheimer\'s')
plt.legend(handles=[blue_patch, teal_patch])
plt.savefig('3D Protein Abundance')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

for i, col in enumerate(target_df.columns[:3]):
    sns.kdeplot(data=target_df[target_df.iloc[:, 3] == 0], x=col, fill=True, label=f'{col} (No Alzheimer\'s)', color='blue', alpha=0.3)
    sns.kdeplot(data=target_df[target_df.iloc[:, 3] == 1], x=col, fill=True, label=f'{col} (Alzheimer\'s)', color='teal', alpha=0.3)

plt.legend()
plt.tight_layout()
plt.savefig('Protein Density Plots')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in target_df.columns[:3]:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data=target_df[target_df.iloc[:, 3] == 0], x=col, fill=True, label='No Alzheimer\'s', color='blue', alpha=0.3)
    sns.kdeplot(data=target_df[target_df.iloc[:, 3] == 1], x=col, fill=True, label='Alzheimer\'s', color='teal', alpha=0.3)
    plt.title(f'{col} - Protein Abundance Distribution')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'{col}.png', bbox_inches='tight') 
    plt.show()
```
